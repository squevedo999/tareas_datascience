- Cree otro dataframe llamado meteoritos_h5. En este dataframe, utilizando la función str. seleccione todos los meteoritos de solamente categoría H5
- Limpie los datos de ambos dataframes, eliminando las entradas donde las masas sean cero, null, NaN, etc.
- Utilizando masks y selecciones, cuente cuántos meteoritos H4 cayeron entre 1960 y 1990.
- Utilizando masks y selecciones, cuente cuántos meteoritos H5 cayeron entre 1960 y 1990.
- Concatene ambos dataframes y grafique la cantidad de meteoritos por clase y por año

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
met = pd.read_csv("meteorite-landings.csv")
met.head()
```

```python
meteoritos_h4 = met[met.recclass.str.startswith('H4')]
meteoritos_h5 = met[met.recclass.str.startswith('H5')]
```

```python
meteoritos_h5
```

```python
meteoritos_h4.recclass.value_counts()
```

```python
meteoritos_h5.recclass.value_counts()
```

```python
meteoritos_h4 = meteoritos_h4[meteoritos_h4.recclass.str.endswith('H4')]
meteoritos_h5 = meteoritos_h5[meteoritos_h5.recclass.str.endswith('H5')]
meteoritos_h4
```

```python
met_h4 =meteoritos_h4.dropna()
met_h5 =meteoritos_h5.dropna()
```

```python
maskh4 = (met_h4['year'] >= 1960) & (met_h4['year']<=1990)
maskh5 = (met_h5['year'] >= 1960) & (met_h5['year']<=1990)
```

```python
h4 = met_h4[maskh4]
h5 = met_h5[maskh5]
```

```python
print(h4.count())
print(h5.count())
```

```python
concatenados = pd.concat([h4,h5], axis=0)
```

```python
concatenados
```

```python
ys1 = h4.groupby('year')['recclass'].count()
ys2 = h5.groupby('year')['recclass'].count()
```

```python
years = np.linspace(1960,1990,31)

plt.plot(years , ys1, label = 'H4')
plt.plot(years, ys2, label = 'H5');

plt.title('Número de meteoritos por año')
plt.legend()
```

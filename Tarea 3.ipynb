```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

datos = sns.load_dataset('penguins')
datos.head()
```

```python
datos
```

Hay 344 filas y 7 columnas

```python
datos.columns
```

```python
datos.species.unique(), datos.island.unique(), datos.sex.unique()
```

No menciono los valores unicos de "bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g", ya que estas columnas están llenas con valores numéricos. 

La columna "species" muestra las diferentes especies de pingunos que se están considerando en el data set, las cuales son: Adelie, Chinstrap, y Gentoo. La columna de "bill length" muestra cuantitativamente el largo del pico de cada pinguino estudiado. 

```python
sns.boxplot(x='bill_length_mm', y = 'species', data = datos,whis=[0, 100], width=.6, palette="vlag")
```

```python
sns.boxplot(x='bill_depth_mm', y = 'species', data = datos,whis=[0, 100], width=.6, palette="vlag");
```

```python
#datos.loc[:,[datos[datos.species.str.startswith("Adelie")],"bill_length_mm"]]
adelie = datos[datos.species.str.startswith("A")]
adelie_bill_length = adelie.loc[:,"bill_length_mm"]

chinstrap = datos[datos.species.str.startswith("C")]
chinstrap_bill_length = chinstrap.loc[:,"bill_length_mm"] 

gentoo = datos[datos.species.str.startswith("G")]
gentoo_bill_length = gentoo.loc[:,"bill_length_mm"] 
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.5, density=False, bins=50)



plt.hist(adelie_bill_length, **kwargs, label = "Adelie")
plt.hist(chinstrap_bill_length, **kwargs, label = "Chinstrap")
plt.hist(gentoo_bill_length, **kwargs, label = "Gentoo")

plt.legend()

```

```python
kwargs = dict(histtype='stepfilled', alpha=0.5, density=False, bins=50)

fig, ax = plt.subplots(figsize=(12, 6))

plt.hist(adelie_bill_length, **kwargs, label = "Adelie")
plt.hist(chinstrap_bill_length, **kwargs, label = "Chinstrap")
plt.hist(gentoo_bill_length, **kwargs, label = "Gentoo")

plt.legend()

ax.text(35,10.2, "Largo más frecuente Adeline")
ax.annotate('Pico más largo', xy=(59.4, 1), xytext=(55, 2),
            arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
datos.corr()
```

Se puede notar que hay un poco de correlación positiva entre el largo de las aletas y el largo del pico debido al valor de 0.656181. Esto mismo se da a menor escala entre la masa y el largo del pico ya que tiene un valor de 0.595110. Asimismo se puede notra una correlación negativa leve entre el largo de la aleta y la profundidad del pico, debido a su valor de -0.583851. Sin embargo la correlación más importante se da entre la masa corporal del pinguino y el largo de su aleta, ya que tiene un valor de correlación de 0.871202.  

```python
sns.heatmap(datos.corr())
```

Del mapa de calor se puede corroborar lo mencinado anteriormente, las correlacion más fuerte se da entre la masa y el largo de las aletas. De la misma forma hay un poco de correlación entre el largo de las aletas y el largo del pico; y entre la masa corporal y el largo del pico también. 

```python
sns.pairplot(datos, hue="species")
```

se puede notar que para el caso de las comparaciones entre el largo de las aletas y la profundidad del pico, las distribuciones son muy similares entre las especies Chinstrap y Adeline. Esto también se observa entre las variables de masa corporal con largo de aleta y con la profundidad del pico. En general las especies Chinstrap y Adeline tienen proporciones similares, mientras que en todos los casos las distribuciones son muy diferentes para la especie Gentoo.
